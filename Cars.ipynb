# Data Pre-processing

Look for the missing values in all the columns and either impute them (replace with mean, median, or mode) or drop them.
Justify your action for this task.     

```python
#Check for null values
car_data.isnull().sum()
```

```python
mean_seats = car_data['Seats'].mean()

# Fill missing values in the "Seats" column with the mean
car_data['Seats'].fillna(mean_seats, inplace=True)

car_data.head(5)
```

```python
#Check for null values
car_data.isnull().sum()
```

Remove the units from some of the attributes and only keep the numerical values

```python
#for Mileage, Engine, Power & New_Price has null values and object data datatype
# Convert the column to strings
car_data['Mileage'] = car_data['Mileage'].astype(str)
car_data['Engine'] = car_data['Engine'].astype(str)
car_data['Power'] = car_data['Power'].astype(str)
car_data['New_Price'] = car_data['New_Price'].astype(str)


# Use the .str accessor and .str.extract to extract the numeric values
car_data['Mileage'] = car_data['Mileage'].str.extract('(\d+\.\d+)', expand=False).astype(float)
car_data['Engine'] = car_data['Engine'].str.extract('(\d+\.\d+)', expand=False).astype(float)
car_data['Power'] = car_data['Power'].str.extract('(\d+\.\d+)', expand=False).astype(float)
car_data['New_Price'] = car_data['New_Price'].str.extract('(\d+\.\d+)', expand=False).astype(float)

#Functio to convert & remove units
def convert_function(text):
    if pd.notna(text):
        matches = re.findall(r'(\d+\.\d+|\d+)', str(text))
        if matches:
            return float(matches[0])
    return None

#Task for Removeing units of Mileage, Engine, Power, New_Price and extracting the Numerical values for the Coloumn
car_data["Mileage"] = car_data["Mileage"].apply(convert_function)
car_data["Engine"] = car_data["Engine"].apply(convert_function)
car_data["Power"] = car_data["Power"].apply(convert_function)
car_data["New_Price"] = car_data["New_Price"].apply(convert_function)
```

```python
car_data.head()
```

```python
#New_Price column empty values fill it up with Median
car_data['New_Price'].fillna(car_data['New_Price'].median(), inplace=True)
```

```python
#Engine column empty values fill it up with 0

car_data['Engine'] = car_data['Engine'].fillna(0)

car_data = car_data.dropna(subset=['Engine'])
```

```python
#Mileage, Power & Seats column empty values fill it up with mean
car_data['Mileage'].fillna(car_data['Mileage'].mean(), inplace=True)
car_data['Seats'].fillna(car_data['Seats'].mean(), inplace=True)
car_data['Power'].fillna(car_data['Power'].mean(), inplace=True)
```

```python
car_data.isnull().sum()
```

Changing the categorical variable  into numerical one hot encoded value for Fuel type and Transmission

```python
car_data = pd.get_dummies(car_data, columns=["Fuel_Type", "Transmission"])
car_data.head()
```

Create one more feature and add this column to the dataset.
For example, you can calculate the current age of the car by subtracting “Year” value from the current year.   

```python
current_year = datetime.datetime.now().year
car_data['Current_Time'] = current_year - car_data['Year']
```
