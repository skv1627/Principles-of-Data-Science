```python
import pandas as pd
data= pd.read_csv("/content/Assignment1/a_raw_data/StudentsPerformance.csv")
data.head()
```

```python
#Check for null values
data.isnull().sum()
```

```python
#Calculating the Average Score
data['Average_Score'] = (data['math score'] + data['writing score'] + data['reading score'])/3
data.to_csv("/content/Assignment1/b_clean_data/CleanData_Student_Performance.csv")
data.head()
```

```python
# visualization 1: Bar plot of gender distribution
import matplotlib.pyplot as mtp

gender_count = data['gender'].value_counts()
mtp.bar(gender_count.index, gender_count.values)
mtp.title('Gender Distribution')
mtp.xlabel('Gender')
mtp.ylabel('Count')
mtp.savefig ('/content/Assignment1/results/gender_distribution.png')
mtp.show()
```

```python
import seaborn as sns
# visualization 2: Box plot of Reading scores by gender
a=sns.boxplot(x='gender', y='reading score',data=data)
mtp.title('Reading Score by Gender')
fig=a.get_figure()
fig.savefig ('/content/Assignment1/results/ReadingScore_by_Gender.png')
```

```python
# visualization 3: Pie chart plot of Race by Average Scores
import plotly.express as px
import matplotlib.pyplot as plt

b = px.pie(data, values="Average_Score", names="race/ethnicity", title='Average Score By Race')
b.show()
```

```python
# visualization 4: Density plot of Reading Scores by Test preparation course
sns.kdeplot(x='reading score', data=data, hue='parental level of education', fill=True, alpha=.3)
plt.title('Reading Scores by Test prep')
plt.xlabel('Reading Score')
plt.savefig ('/content/Assignment1/results/Reading_score_by_ParentalLevelEducation_course.png')
plt.show()
```

```python
plt.hist([data['writing score'],data['math score'],data['reading score']], bins=20,  color=['red','green','blue'], edgecolor='black')
plt.xlabel('All 3 types of score')
plt.ylabel('Values')
plt.title('Histogram of All 3 types of score')
plt.show()
plt.savefig('/content/Assignment1/results/histograh.PNG')
```
