```python
#RAW_DATA_COLLECTION
#importing raw data
import pandas as pd
data = pd.read_csv("/content/Assignment1/Women_Frailty/a_raw_data/rawdata_women_fratility_data.csv")
data.head()
```

```python
#CLEANING_DATA
#checking for null values
data.isnull().sum()
```

```python
#converting Frailty Attribute to binary values
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
data['Frailty']=le.fit_transform(data['Frailty'])
data.head()
```

```python
#Exporting the clean data file to b_clean_data folder
data.to_csv('/content/Assignment1/Women_Frailty/b_clean_data/cleandata_women_frailty_data.csv')
```

```python
#DATA_ANALYSIS
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#Visualizing the relation between Weight and Grip_strength

#Create a bar plot of 'Weight' by 'Grip Strength'
a = sns.barplot(data, x='Weight', y='Grip_Strength')
fig = a.get_figure()
fig.savefig("/content/Assignment1/Women_Frailty/results/Weight_GripStrength_relation.png")
```

Here it is evident that even the women who has much weight doesn't posses great grip strength.

```python
# Create a box plot of 'Grip strength' by 'Frailty'
a = sns.boxplot(data, x='Frailty', y='Grip_Strength')
fig = a.get_figure()
fig.savefig("/content/Assignment1/Women_Frailty/results/Frailty_GripStrength_relation.png")
```

```python
#Create a scatter plot of 'Weight' vs. 'Height'
a=sns.scatterplot(data, x='Height', y='Weight', hue='Frailty')
fig = a.get_figure()
fig.savefig("/content/Assignment1/Women_Frailty/results/Height_Weight_Frailty_relation.png")
```
